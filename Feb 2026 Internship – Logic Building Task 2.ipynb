```python
def Unique_Words_in_a_Sentence(s):
  word= s.split()
  unique=set(word)
  print("unique words count:",len(unique))
  print("unique words:",unique)
Unique_Words_in_a_Sentence("python is easy and python is powerful")
```

```python
def highest_salary(emp):
  highest_name=max(emp,key=emp.get)
  print("highest salary :",highest_name,"-",emp[highest_name])
emp={
    "ravi":7600,
    "sana":6000,
    "kiran":9000
}
highest_salary(emp)
```

```python
def find_max_mini(numbers):
  maximum=numbers[0]
  minimum=numbers[0]
  for i in numbers:
    if i >maximum:
      maximum=i
    if i<minimum:
      minimum=i
  print("list",numbers)
  print("max",maximum)
  print("mini",minimum)
find_max_mini([43 ,23 ,54 ,34 ,76])
```

```python
def count_above_1000(prices):
  count=0
  for i in prices:
    if i>1000:
      count+=1
  print("product above 1000",count)

count_above_1000([345 ,6000 ,76 ,500 ,457])
```

```python
def attendance_percentage(attendence):
  present=attendence.count("p")
  total=len(attendence)
  percentage=(present/total)*100
  print("Attendence percange:",percentage)
attendance_percentage(["p","p","a","p"])
```

```python
def unique_phone_numbers(phone_numbers):
  unique=set(phone_numbers)
  print("unique phone number:",unique)
unique_phone_numbers([970765567,795265920,9278409754])
```
