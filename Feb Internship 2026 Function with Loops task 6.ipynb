```python
def post_engagment_analyzer(like_list):
  total_likes=0
  for like in like_list:
    total_likes +=like
  if total_likes >=1000:
    status="viral post"
  else:
    status='NOrmal engagment'
  print("total likes:",total_likes)
  print("post status",status)
likes=[200,300,400,500]
post_engagment_analyzer(likes)
```

```python
def attendance_checker(attendance_list):
  total_classes=len(attendance_list)
  present=0
  for status in attendance_list:
    if status.upper()=="P":
      present+=1
  percentage=(present/total_classes)*100
  if percentage>=75:
    result="eligible"
  else:
    result="not eligible"
  print("attendance percantage",percentage)
  print("exam eligibility:",result)
attendance_checker(["a","p" ,"p" ,"p" ,"a"])
```

```python
def electricity_bill(units):
  bill=0
  if units<=100:
    bill=units*3
  elif units<=200:
    bill=(100*3)+((units-100)*5)
  else:
    bill=(100*3)+(100*5)+((units-200)*7)
  if bill <=1500:
    status="moderate usage"
  else:
    status="high usage"
  return bill , status
bill, usage=electricity_bill(250)
print("total bill:",bill)
print("usage level",usage)
```
