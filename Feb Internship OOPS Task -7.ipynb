```python
class  Smart_Light_Controller:

  def __init__(self,lt_name,action):
    self.lt_name=lt_name
    self.action=action
  def display(self):
    print(f"light name:{self.lt_name}")
    print(f" action:{self.action}")
  def smart_home_automation(self):
    if self.action.lower()=="on":
      print("ligth is turned on")
      self.display()
    elif self.action.lower()=="off":
      print("light is turned off")
      self.display()
lt_name=input("enter the name of the light")
action=input("enter the action of the light")
obj=Smart_Light_Controller(lt_name,action)
obj.smart_home_automation()


```
