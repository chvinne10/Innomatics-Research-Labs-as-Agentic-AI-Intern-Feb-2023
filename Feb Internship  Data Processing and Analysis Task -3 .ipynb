```python
import string
query="buy mobile hone buy phone online"
query=query.lower()
query=query.translate(str.maketrans('','',string.punctuation))
words=query.split()
frequency={}
for  word in words:
  frequency[word]=frequency.get(word,0)+1
rep_key={word : count for word, count in frequency.items() if count>1}
print(rep_key)
```

```python
sensor_readings=[1 ,2 ,3 ,4 ,5 ,7 ,6]
valid_readings=[]
for hour , reading in enumerate(sensor_readings):
  if reading%2==0:
    valid_readings.append((hour,reading))
print("valid sensor reading (hou,value):")
print(valid_readings)
```

```python
email=["ravi@gmail.com","anita@yohoo.com","suresh@gmail.com","meena@yahooo.com"]
domain_count={}
for email in email:
  domain=email.split("@")[1]
  domain_count[domain]=domain_count.get(domain,0)+1
total_emails=len(email)
for domain,count in domain_count.items():
  percentage=(count/total_emails)*100
  print(f"{domain}:{int(percentage)}%")


```

```python
sales=[1200,1500,900,2200,1400,3000]
average_sales=sum(sales)/len(sales)
thresdhold=average_sales*1.3
for days,sale in enumerate(sales,start=1):
  if sale>thresdhold:
    print(f"day{days}:sales")

```
