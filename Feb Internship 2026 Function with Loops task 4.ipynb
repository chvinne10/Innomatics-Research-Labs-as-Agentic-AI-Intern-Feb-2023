```python
def validate_a_mobile_recharge_request(amount):
  while amount!=0:
    if amount<50:
      print( "Recharge amount must be ≥ ₹50")
      amount=int(input("enter valid amount:"))
    if amount not in [199, 299, 399, 599]:
        print("Valid plans are not  exist enter valid plans")
        amount=int(input("enter valid plan"))
    else:
      return " mobile recharge request is succussful"
amount=int(input("enter the mobile recharge plan:"))
print(validate_a_mobile_recharge_request(amount))
```

```python
def Inventory_Reorder_Alert_System(dictionary):
  stock=0
  for i in dictionary.values():
    stock=stock+i
  if stock<15:
    return "Reorder Alert"
  else:
    return f"stock ok,quantity of stock{stock}"
dictionary={}
number=int(input("enter the no. of items:"))
while number !=0:
  keys=input("enter the product name:")
  value=int(input("enter the stock quantities:"))
  dictionary[keys]=value
  number=number-1
print(Inventory_Reorder_Alert_System(dictionary))
```

```python
def Student_Result_Processing_System(lst):
  avg=0
  sum=0
  for i in lst:
    sum=sum+i
  avg=sum/len(lst)
  if avg>=50:
    print(f"the student pass {avg}")
  else:
    print(f"the student fail {avg}")
lst=[]
num=int(input("enter the no of subject of student"))
i=0
while i<num:
  a=int(input(f"enter the student maks subject:{i+1}:"))
  lst.append(a)
  i=i+1
Student_Result_Processing_System(lst)

```

```python
def Employee_Attendance_Eligibility_Checker(lst):
  count=0
  for i in lst:
    if i.upper()=="P":
      count=count+1
  if (count/len(lst))*100>=75:
      return   "Eligible"
  else:
      return " Not Eligible"
lst=[]
i=0
a=int(input( "enter total number of days  for attendance calculate:" ))
while i<a:
    attendance=input("enter only P/A for absent and present:")
    lst.append(attendance)
    i=i+1
print(Employee_Attendance_Eligibility_Checker(lst))




```

```python

def  Password_Strength_Checker(password):
  if len(password)<8:
    return "weak password"
  has_digit=False
  has_special=False
  for i in password:
    if i.isdigit():
       has_digit=True
    if i in "@#$":
       has_special=True
  if has_digit and has_special == True:
    print("strong password")
  else:
    print("weak password")
password=input(" enter the password")
Password_Strength_Checker(password)


```
