```python
def Login_Check(user,passw):
  if user=="admin" and passw=="1234":
    return "Login Successful"
  else:
    return "Invalid Credentials"
user=input("Enter Username: ")
passw=input("Enter Password: ")
print(Login_Check(user,passw))
```

```python
def Pass_Fail_Analyzer(lst):
  Pass=0
  Fail=0
  for i in lst:
    if i>=50:
      Pass=Pass+1
    else:
      Fail=Fail+1
  print("Pass: ",Pass)
  print("Fail: ",Fail)
marks = [45, 78, 90, 33, 60]
Pass_Fail_Analyzer(marks)
```

```python
def Simple_Data_Cleaner(lst):
  clear_list=[]
  for i in lst:
    i=i.strip()
    i=i.lower()
    clear_list.append(i)
  return clear_list
names = [" Alice ", "bob", " CHARLIE "]
print(Simple_Data_Cleaner(names))
```

```python
from re import M
def Message_Length_Analyzer(lst):
  flag=10
  for i in lst:
    if len(i)>=flag:
      print(f"messages longer than 10 character:'{i}'")
    else:
      print(f"the length of each message is:'{i}'")
messages = ["Hi", "Welcome to the platform", "OK"]
Message_Length_Analyzer(messages)
```
